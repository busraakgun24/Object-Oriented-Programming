```python
class Book:#Bir kitap nesnesini temsil etmek için gerekli özellikleri ve bu özelliklere
#erişim/değiştirme yöntemlerini (metotlarını)içeren Book adında bir sınıf oluşturduk
    def __init__(self, kitap_adi, yazar, sayfa_sayisi, yil):#yapıcı metot.Bu metot bir book nesnesi oluşturulurken çağırılır ve nesneye ilk değernini atar.
    #self:nesnenin kendisini temsil eden zorunlu bir parametredir.
        self.__kitap_adi = kitap_adi#her bir book nesnesine ait olan özellikler __(private) olarak doğrudan erişime izin vermeyecek şekilde oluşturuldu
        self.__yazar = yazar##sağdaki ifadenin (değişkenin) değeri soldaki ifadeye atanır
        self.__sayfa_sayisi = sayfa_sayisi
        self.__yil = yil
        self.__available = True  # Mevcut mu değil mi? Ödünç verildiyse False olacak.Başlangçta her kitap mevcut olarak varsayılıyor.

    # Getter - Setter Erişimci ve düzenleyici metotlar bu metotlar nesne özelliklerine güvenli bir şekilde dışarıdan erişmek ve onları değiştirmek için kullanılır.Kapsülleme
    def get_kitap_adi(self):#get metotları özel niteliklerin değerlerini döndürür.Okuma işlemi
        return self.__kitap_adi

    def set_kitap_adi(self, yeni_ad):#set metotları niteliklerin değerlerini değiştirmeye izin verir. Genellikle veri doğrulaması (validation)içerirler.
        if isinstance(yeni_ad, str) and len(yeni_ad.strip()) > 1:#isinstance(yeni_ad,str): Yeni adın bir string (metin) olup olmadığını ve
        #strip() ile boşlukları atıldıktan sonra 1 karakterden uzun olup olmadığını kontrol eder yani bir isim girilmesini zorunlu kılar
            self.__kitap_adi = yeni_ad.strip()#şart doğruysa girilen yeni kitap ismi kitap_adi adlı değişkene atanır

    def get_yazar(self):
        return self.__yazar

    def get_sayfa_sayisi(self):
        return self.__sayfa_sayisi

    def get_yil(self):
        return self.__yil

    def get_available(self):
        return self.__available

    def set_available(self, durum: bool):#kitabın uygunluk durumunu değiştirmek için kullanılır
        self.__available = bool(durum)

    # Polimorfizm: farklı sınıflarda farklı display_info
    def display_info(self):#kitabın adı ve yazarı bilgilerini ekrana yazar
        return f"Kitap: {self.__kitap_adi} | Yazar: {self.__yazar}"

    def __str__(self):#Kitabın uygunluk durumuna göre yazdırma işlemi için kullanılacak metot. string(metin)döndürür
        durum = "Uygun" if self.__available else "Ödünçte"
        return (f"{self.__kitap_adi} - {self.__yazar} | "
                f"{self.__sayfa_sayisi} sayfa | {self.__yil} | {durum}")

class User:#User(kullanıcı) sınıfı tanımlandı
    def __init__(self, username):#yapıcı metot özel tanımlı kullanıcı adı değişkenine ilk değerini atar
        self.__username = username
        self.borrowed_books = []#kullanıcının ödünç aldığı (borrowed) kitap nesnelerini saklayan bir liste oluşturur.Bu liste başta boştur.

    def get_username(self):
        return self.__username

    def set_username(self, name):
        if isinstance(name, str) and len(name.strip()) > 1:#yeni kullanıcı adı atanmadan önce metin olup olmadığı ve boşluklar silindikten
        # sonra 1 karakterden uzun olup olmadığını kontrol eder
            self.__username = name.strip()

    def display_info(self): #Sadece kullanıcı adını içeren bir metin döndürür
        return f"Kullanıcı: {self.__username}"

    def __str__(self):#nesnenin metin temsilini döndürürken kullanıcının ödünç aldığı tüm kitapların adlarını listeler
        borrowed = ", ".join([b.get_kitap_adi() for b in self.borrowed_books]) or "Yok"#ödünç alınan her b kitabı içinget_kitap_adi()metodunu çağırarak adını alır
        #.join(...):kitap adlarını virgül ve boşluk ile birleştirir.
        return f"User({self.__username}) | Ödünçte: {borrowed}"


class Admin(User):#admin:Yönetici sınıfı tanımlandı ve Admin sınıfı User adlı sınıfın tüm özelliklerini ve metotlarını miraz aldı.Kalıtım.Ebeveyn:User sınıfıdır.
    def __init__(self, username, level="Supervisor"):
        super().__init__(username)#miras alınan üst sınıfın yapıcı metodu çağırılır.Username niteliği User sınıfı tarafından başlatılacak
        self.level = level#Yöneticiye özel olarak bir seviye (varsayılanı "Supervisor") niteliği ekler

    def display_info(self):#mirasw aldığı get_username() metodunu kullanarak kullanıcı adını ve yöneticinin seviyesini içeren bir metin döndürür
        return f"Admin: {self.get_username()} | Level: {self.level}"

    def __str__(self):#yöneticiye özel metin temsilini döndürür
        return f"Admin({self.get_username()}, {self.level})"


class Library:#library sınıfı oluşturuldu. Bu sınıf tüm sistemi (kitaplar ve kullanıcılar) yöneten ana yapıdır.
    def __init__(self, name):#Constructor
        self.name = name#Kütüphanenin adını tutar
        self.books = []#liste kullanımı.Kütüphanedeki tüm Book nesnelerini saklayacak bir sözlük oluşturur.
        #Anahtar olarak kitabın adının küçük harfe çevrilmiş hali tutulur
        self.users = {}#Kütüphanedeki tüm User ve Admin nesnelerini saklayacak bir sözlük oluşturur
    #Bu metotlar,kütüphane sistemindeki kitap ekleme,silme, listeleme ve ödünç verme işlemlerini yönetir
    def add_book(self, book: Book):
        if isinstance(book, Book):#isinstance tip kontrolü yapar.Eklenecek olan objenin Book sınıfı nesnesi
        # olup olmadığı kontrol edilir.Yanlış nesne eklenmesini engeller
            self.books.append(book)
            return True
        return False

    def remove_book(self, book_name: str):#Kütüphaneden bir kitabı adıyla kaldırır.
        target = book_name.strip().lower()#kitabın adı küçük harfe çevrilir ve fazladan boşluklar silinir ve sözlük.
        # anahtarı(target)oluşturulur. Bu aramaları tutarlı hale getirir
        for b in self.books:#kitap sözlüğünün her elemanını(nesne:b) gezecek olan döngü kuruldu.
            if b.get_kitap_adi().strip().lower() == target:#target kitabın arama için standardize edilmiş(hedef) halini tutar
                # sadece silmeye izin ver: eğer ödünçteyse önce uyar.
                if not b.get_available():#kitap mevcut değilse.
                    return "ödünçte"
                self.books.remove(b)
                return True
        return False

    def list_books(self):
        if not self.books:# liste boş mu? Kontrol eder
            print("Kütüphane boş.")
            return
        for idx, b in enumerate(self.books, start=1):#idx sıra numarasını 1'den başlayarak, b kitap nesnesini tutar. Kitap nesneleri üzerinde döngü kurar.
            print(f"{idx}. {b}")

    def loan_book(self, book_name: str, user: User):# Bir kitabı ödünç verme işlemini gerçekleştirir.
        target = book_name.strip().lower()#Küçük harfleşden oluşan boşlukları atılmış anhatar hazırlanır.
        for b in self.books:#kütüphaneedeki kitaplaqr üzerinde dolaşacak dögü kuruldu.
            if b.get_kitap_adi().strip().lower() == target:#eşleşme kontrolü yapılır
                if not b.get_available():# b uygun değilse çalışacak olan şart yapısı
                    return "not_available"
                b.set_available(False)#durum güncellemesi yapar.
                user.borrowed_books.append(b)#Kittabı kullanıcının ödünç aldığı kitaplar listesine ekler.
                return True
        return False

    def return_book(self, book_name: str, user: User):#Kitabın geri alınması işlemini yapar.
        target = book_name.strip().lower()#İade alınacak kitabın adını standardize eder.
        for b in user.borrowed_books:#Ödünç alınan kitaplar listesinde dönecek döngü kurulur.
            if b.get_kitap_adi().strip().lower() == target:# eşleşme olursa çalışacak şart yapısı.
                b.set_available(True)#Kitabın durmu güncellendi artık uygun/mevcut.
                user.borrowed_books.remove(b)#İade edilen kitabı kullanıcının ödünç alığı kitaplar listesinden kaldırır.
                return True
        return False


def main():#Kullanıcı arayüzünü içeren fonsiyon tanımlaması yapıldı. Bu fonksiyon program çalıştığında
#ilk çağırılan ve kütüphane sistemini başlatan ana döngüdür.
    kutuphane = Library("Huzur Library")#Library sınıfından  bir örnek(obje/nesne) oluşturuldu. Adı Huzur Library kondu(değer gönderme)
    admin = Admin("Müdür")# Admin sınıfından bir nesne oluşturuldu. İsmi Müdür kondu.
    kutuphane.users[admin.get_username().lower()] = admin#Oluşturulan admin nesnesi, kütüphanenin kullanıcılar sözlüğüne ekledik.

    print("Kütüphane sistemi başlatıldı...")

    while True:#Kullanıcıya istediği işlemi seçme seçeneği sunan menü tasarlandı.Sonsuz döngü
        print("\n---- MENÜ ----")
        print("1 - Kitap Ekle")
        print("2 - Kitap Sil")
        print("3 - Kitapları Listele")
        print("4 - Ödünç Kitap Ver")
        print("5 - Kitap Geri Al")
        print("6 - Çıkış")

        secim = input("Seçiminiz: ").strip()

        if secim == "1":
            ad = input("Kitap Adı: ").strip()#Veri girişi
            yazar = input("Yazar: ").strip()#Veri girişi
            sayfa = input("Sayfa Sayısı: ").strip()
            yil = input("Yıl: ").strip()

            if not (sayfa.isdigit() and yil.isdigit()):#Sayfa sayısı ve yıl değişkenlerinin sadece rakamlardan oluşup oluşmadığını kontrol eder isdigit()
            #Opsinel olarak yılın belli sayı aralığında girilmesi de kontrol edilebilirdi.Aynı doğrultuda sayfa sayısının sadece pozitif tam sayıları
            #kabul etmesi sağlanabilirdi. trp ve except blokları ile

                print("Sayfa sayısı ve yıl rakam olmalı.")
                continue

            kitap = Book(ad, yazar, int(sayfa), int(yil))
            kutuphane.add_book(kitap)#Kitap kütüphaneye eklendi.
            print("Kitap başarıyla eklendi.")

        elif secim == "2":
            sil = input("Silinecek kitap adı: ").strip()
            res = kutuphane.remove_book(sil)#Kütüphanenin remove metodunu çağırır vesonucu res (sonuc) deeğişkenine atar
            if res is True:
                print("Kitap silindi.")
            elif res == "ödünçte":#Ödğnçte ise uyarı verir ve silme işlemi yapmaz
                print("Kitap şu anda ödünçte, önce iade edilmeli.")
            else:
                print("Kitap bulunamadı.")

        elif secim == "3":# metot çağırılarak kitaplr listelenir.
            kutuphane.list_books()

        elif secim == "4":
            kitap_ad = input("Ödünç verilecek kitap adı: ").strip()# ödünç alınacak kitabın adı girilir.
            kullanici_ad = input("Kullanıcı adı: ").strip()#kitabı alacak kişinin adı girilir.
            key = kullanici_ad.lower()#Büyük-küçük harf uyumuna takılmaz.Tüm harfler küçük yapıldı.
            if key not in kutuphane.users:#key kütüphane kullanıcıları sözlüğünde (kutuphane.users)var mı diye kontrol eder
                kutuphane.users[key] = User(kullanici_ad)#
            user = kutuphane.users[key]#kullanıcı bulunduysa ilgili user veya Admin nesnesini alır.
            res = kutuphane.loan_book(kitap_ad, user)#kütüphanenin loan_book metodunu çağırır vesonucu res değişkenine atar.
            if res is True: #sonuç doğruysa
                print(f"{user.get_username()} kitabı ödünç aldı.")
            elif res == "not_available":
                print("Kitap zaten ödünçte.")
            else:
                print("Kitap bulunamadı.")

        elif secim == "5":
            kitap_ad = input("Geri alınacak kitap: ").strip()
            kullanici_ad = input("Kullanıcı adı: ").strip()
            key = kullanici_ad.lower()
            if key not in kutuphane.users:
                print("Böyle bir kullanıcı yok.")
                continue
            user = kutuphane.users[key]
            res = kutuphane.return_book(kitap_ad, user)
            if res is True:
                print("Kitap kütüphaneye geri alındı.")
            else:
                print("Bu kullanıcıda belirtilen kitap görünmüyor.")

        elif secim == "6":
            print("Sistem kapanıyor.")
            break
        else:
            print("Geçerli bir seçim yap.")


if __name__ == "__main__":^program başlatma kontrolü
    main()#Fonksiyon çağrısı yapıldı. main fonksiyonunun parametresi olmadığı için çağırırken parametre göndermedik.
```
