```python
#kütüphane verilerini tutacak global liste tanımlanmıştır.
#Her kitap bir liste olarak ana listeye eklenecektir.
# Dict de kullanabilirdim ama dict genellikle OOP yaklaşımında bir nesnenin
# özelliklerini tutmaya çok benzer bi mantıkla çalışır.
kutuphane_sistemi = [] #kütüphanedeki tüm kitap verilerini tutacak olan boş ana listeyi tanımlar. Tüm işlemler bu liste üzerinden yapılacak
```

```python
def kitap_ekle():# kitap_ekle fonksiyonu başlıyor.
  """Yeni bir kitap bilgisi alır ve onu kütüphane listesine ekler. En önemlisi, aynı isimde kitap eklenmesini kesinlikle engeller."""
  print("\n---Kitap Ekleme---")
  # Bu döngü, kullanıcı geçerli ve daha önce eklenmeyen bir isim girene kadar devam edecek.
  while True:
    kitap_adi=input("Kitap Adı: ").strip() # Kullanıcıdan kitap adını alıyoruz ve .strip() built in metodu ile gereksiz boşlukları temizliyoruz.
    if not kitap_adi: # Eğer giriş boşsa (False ise)
      print("HATA: Kitap Adı boş bırakılamaz. Lütfen bir kitap adı giriniz.")
      continue #Döngünün başına dönüyoruz.
    #Benzersizlik Kontrolü. Harfe duyarsız arama yapılacak.
    kitap_adi_lower = kitap_adi.lower() # Aramayı kolaylaştırmak için girişi küçük harfe çeviriyoruz.
    kitap_mevcut = False # Bir bayrak(flag) değişkeni tanımlıyoruz.
    # Kütüphane listesindeki tüm kayıtları tek tek kontrol etme döngüsü.Duyarsız karşılaştırma yapılacak.
    for kitap in kutuphane_sistemi:
        if kitap[0].lower() == kitap_adi_lower:
            kitap_mevcut = True
            break # Kitap bulunduğu anda döngü duracak.
    if kitap_mevcut:
        print(f"HATA: '{kitap_adi}' adlı kitap zaten kütüphanede kayıtlı.")
        continue # Hatalı isim girildiği için tekrar ad sormak üzere döngü başına dönüyoruz.
    break
  # Kitap Adı'na benzer şekilde, yazar adının boş bırakılmasını engelleyecek yapı.
  while True:
    yazar=input("Yazar Adı:").strip() #Kullanıı girişi: yazar adını alıyoruz.
    if yazar: # Eğer yazar adı doluysa
      break
    print("HATA: Yazar adı boş bırakılamaz. Lütfen bir yazar adı giriniz.")
  # Sayfa sayısı için 'try-except' kullanarak kullanıcıdan tam sayı almayı garanti ediyoruz.
  while True:
    try:#Hata çıkma olasılığı olan kodu dener.
      sayfa_sayisi=int(input("Sayfa Sayısı:").strip())
      if sayfa_sayisi<=0: # Mantıksal kontrol: Sayfa sayısı kesinlikle pozitif olmalı.
        print("HATA: Sayfa sayısı 0'dan küçük veya eşit olamaz.")
        continue
      break
    except ValueError: # Kullanıcı harf veya sembol girdiğinde yakalıyoruz.
      print("HATA: Geçersiz giriş. Sayfa sayısı bir tam sayı olmalıdır.")
  while True:
    try:
      basim_yili=int(input("Basım Yılı:").strip())#Kullanıcıdan giriş alıyoruz.Fazlalığını atıp basim_yili adlı değişkene atıyoruz.
      if basim_yili<1000 or basim_yili>2025: # Yıl aralığı kontrolü yapıyoruz.
        print("HATA: Geçersiz yıl girdiniz. Lütfen mantıklı bir yıl giriniz: (1000-2025 arası).")
        continue
      break
    except ValueError:
      print("HATA: Geçersiz giriş. Basım yılı bir tam sayı olmalıdır.")
  durum='Mevcut' # Yeni eklenen her kitap varsayılan olarak 'Mevcut' değerini alacak.
  # Tüm verileri listeler listesi yapısına uygun sırada topladık: [Ad, Yazar, Sayfa, Yıl, Durum]
  yeni_kitap=[kitap_adi, yazar, sayfa_sayisi, basim_yili, durum]
  kutuphane_sistemi.append(yeni_kitap) # Oluşturulan listeyi Python'un .append() built in (Tr:Kurulu/yerleşik) metodu ile global kütüphane listesine ekler.
  print(f"\n '{kitap_adi}' adlı kitap başarıyla kütüphane sistemine eklendi. Kayıt tamamlandı.")
```

```python
def kitap_sil():#kitap_sil adlı fonksiyon tanımlanıyor
  """Kütüphane listesinden bir kitabı adına göre siler."""
  print("\n ---Kitap Silme:---")
  if not kutuphane_sistemi:#kütüphane listesinin boş olup olmnadığını kontrol eder.
    print("Kütüphane boş. Silinecek kitap yok.")
    return
  adi_silinecek=input("Silmek istediğiniz kitabın adını giriniz:").strip()#silinecek kitabın adı alınır ve boşluklar atılır.
  kitap_bulundu= False# kitabın bulunup bulunmadığını takip etmek için bir boolean (bayrak)değişken tanımlanır. Varsayılan False atanır.
  #listenin üzerinde döneceğiz, adı eşleşen elemanı bulacağız.
  for i in range(len(kutuphane_sistemi)):#listedeki her elemanı gezen döngü
  #kutuphane_sistemi[i][0]->Kitabın adı
    if kutuphane_sistemi[i][0].lower()==adi_silinecek.lower():#büyük-küçük harfe duyarsız arama işlemi yapıyoruz
      del kutuphane_sistemi[i]# eşleşirse o kitaba ait alt listeyi ana listeden siler
      print(f"\n '{adi_silinecek}' adlı kitap başarıyla silindi.")
      kitap_bulundu=True
      break# arama işlemi bitti
  if not kitap_bulundu:
    print(f"\n'{adi_silinecek}'adlı kitap bulunamadı.")
```

```python
def kitaplari_listele():# kitapları listeleyecek fonksiyon tanımlandı.
  """Kütüphane listesindeki tüm kitapları listeler."""
  print("\n ---Kütüphane Listesi:--- ")
  if not kutuphane_sistemi:#kütüphane llistesinin boş olup olmadığını kontrol eder
    print("Kütüphane boş.")
    return
  print("No| Kitap Adı      | Yazar          | Sayfa | Durum")
  for i, kitap in enumerate(kutuphane_sistemi):#kutuphane_sistemi adlı listedeki her alt listeyi ve onun sırasını almak için döngü başlatır
    print(f" {i+1} | {kitap[0]} | {kitap[1]} | {kitap[2]} | {kitap[3]} | {kitap[4]} ")
```

```python
def odunc_kitap():#odunc_kitap adlı fonksiyon tanımlandı
  """Kütüphanedeki bir kitabın durumunu 'Ödünçte' olarak günceller."""
  print("\n---Ödünç ver---")
  if not kutuphane_sistemi:# Kütüphane listesinin boş olup olmadığını kontrol eder.
    print("Kütüphane boş. Ödünç verilecek kitap yok.")
    return
  ad_odunc = input("Ödünç alınacak kitabın adını giriniz:").strip()# Ödünç alınmak istenen kitabın adı kullanıcıdan alınır.
  kitap_bulundu = False # Kitabın bulunup bulunmadığını izler. Varsayılan False atanır.
  for kitap in kutuphane_sistemi:# kutuphane_sistemi adlı listedeki her alt listeyi (kitabı) dolaşacak döbgü
    # Harf Duyarsız Arama Kontrolü:
    # kitap[0] (kitap adı) ile kullanıcının girdiği ad (ad_odunc) küçük harfe çevrilerek karşılaştırılır.
    if kitap[0].lower() == ad_odunc.lower():
      kitap_bulundu = True # Kitap listede bulundu
      # Durum İndeksi Kontrolü: Kitabın durumu listenin 4. indeksinde (5. eleman) tutulur.
      if kitap[4] == 'Mevcut':# Kitabın durumu (mevcut olup olmadığı) kontrol edilir..
        kitap[4] = 'Ödünçte'# Mevcutsa durumunu ödünçte olarak değiştirir.
        print(f"'{ad_odunc}' ödünç verildi.")
      else:# Mevcut değilse (yani durumu 'Ödünçte' ise)
        print(f"'{ad_odunc}' şu anda başka bir kullanıcı tarafından ödünç alınmıştır.")

      # İşlem yapıldığı için döngüyü ve fonksiyonu sonlandır
      return
  # Eğer döngü biter ve kitap hala bulunamadıysa (kitap_bulundu hala False)
  if not kitap_bulundu:
    print(f"'{ad_odunc}' adlı kitap kütüphanede bulunamadı.")
```

```python
def geri_al():#geri_al adında fonksiyon tanımladık
  """Ödünç alınmış bir kitabın  durumunu 'Mevcut' olarak günceller."""
  print("\n ---Kitabı Geri Al---")
  if not kutuphane_sistemi:#kütüphane listesinin boş olup olmadığı kontrol edilir
    print("Kütüphane boş.")
    return
  ad_geri=input("Geri vermek istediğiniz kitabın adı: ").strip()#Geri verilecek kitabın adını kullanıcıdan alır ve boşluklaarı siler
  kitap_bulundu = False #Kitabın bulunup bulunmadığını izlmek için boolean değişkeni tanımlanır ve varsayılan False atanır

  for kitap in kutuphane_sistemi:#kutuphane_sistemi listesinin her elemanını gezecek döngü kuruldu.
    if kitap[0].lower()==ad_geri.lower():#geri verilecek kitap listede bulunana kadar kontrol edecek ve bu işlem küçük-büyük harfe duyarsızdır.
      kitap_bulundu = True #Kitap bulundu.
      if kitap[4]=='Ödünçte':#Kitabın durumunun 'Ödünçte' olup olmadığını kontrol eder.
        kitap[4]='Mevcut'#Eğer ödünçte ise durumunu mevcut olarak günceller
        print(f"'{ad_geri}' geri alındı ve kütüphaneye kaydedildi.")
      else:
        print(f"'{ad_geri}' zaten kütüphanede(mevcut).")
      return

  if not kitap_bulundu: #Döngü tamamlandı ama kitap bulunamadı ise
    print(f"'{ad_geri}' ödünçte olan kitaplar arasında bulunamadı.")
```

```python
def menuyu_goster():
  """Kullanıcıya ana menüyü gösterir"""
  print("\n---Kütüphane Sistemine Hoşgeldiniz ---")
  print("1-Kitap Ekle, 2-Kitap Sil, 3-Listele, 4- Ödünç Ver, 5-Geri Al, 6-Çıkış")#Kullanıcıya doğru seçim yapabilemesi için bilgilendirme yapar
```

```python
def main():#main adlı ana fonksiyon tanımlanır
  """Programın ana döngüsünü çalıştırır."""
  while True:#Program kapatılana kadar sürekli çalışmasını sağlayan sonsuz döngü başlatılır.
    menuyu_goster()#menuyu_goster fonksiyonu çağırılır ve menü ekrana yazdırılır.
    secim=input("Seçim(1-6) yapınız.").strip()#Kullanıcıdan bir menü numarası girmesini ister ve fazla boşluğu atar.
    if secim=='1':#seçim 1 ise...bu kontrol switch case ile de yapılabilir
      kitap_ekle()#kitap_ekle fonksiyonu çağırılır.
    elif secim=='2':#seçim 2 ise...
      kitap_sil()#kitap_sil fonksiyonu çağırılır.
    elif secim=='3':
      kitaplari_listele()#kitapları_listele fonksiyonu çağırılır.
    elif secim=='4':
      odunc_kitap()#odunc_kitap fonksiyonu çağırılır.
    elif secim=='5':
      geri_al()#geri_al fonksiyonu çağırılır.
    elif secim=='6':
      print("Çıkış yapılıyor...")
      break
    else:
      print("Geçersiz seçim yaptınız.")
if __name__=="__main__":#Bu ifade, kodun bir modül olarak değil, doğrudan çalıştırıldığında main() fonksiyonunu çağırarak Programın çalışmaya başlamasını sağlar.

  main() #main() fonksiyonu çağırılarak programı başlatan standart Python yapısı.
```
